```python
import pandas as pd
from datetime import timedelta, date
```

```python
df = pd.read_csv('../backend/data/dsa_data.csv')
df.dtypes
```

```python
df['Datetime'] = pd.to_datetime(df['Datetime'])
df['Datetime'] = df['Datetime'] + timedelta(hours=8)
df
```

```python
df['Datetime'].dt.date.value_counts().sort_index()
```

```python
(df['Datetime'].dt.date == date(2023,2,23)).sum()
```

## Dummy for Normal Period

```python
# select 23 Feb 2023
normal_df = df[df['Datetime'].dt.date == date(2023,2,23)]
normal_df = normal_df[['Datetime', 'Direction']]
normal_df.to_csv('../backend/data/dummy_normal_period.csv')
```
